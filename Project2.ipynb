Comparison operators: equals, not equals, greater than, less than

```python
Comparison operators in Python are used to compare two values and return a boolean result of either True or False. These include operators such 
as == to check if values are equal, != to check if they are not equal, > and < to determine if one value is greater than or less than another, and 
>= and <= to check for greater than or equal to or less than or equal to conditions. These operators are essential for making decisions in programs,
enabling conditional statements and loops to execute different actions based on whether these comparisons are true or false.
x = 10
y = 20
print(x == y)  # False
print(x != y)  # True
print(x > y)   # False
print(x < y)   # True
print(x >= 10) # True
print(y <= 20) # True
```

For loops: Repeating tasks multiple times automatically

```python
For loops in Python are used to repeat a block of code multiple times automatically by iterating over a sequence such as a list,
string, or range of numbers. The syntax involves the for keyword followed by a variable name, the in keyword, and the sequence to loop through. 
In each iteration, the loop variable takes on the next value from the sequence, and the code block inside the loop executes once per item. For
example, you can loop over a list of fruits to print each fruit's name, or use range() to run a loop a specific number of times, such
as counting from 0 to 5. For loops simplify repetitive tasks by automating them, eliminating the need to write duplicate code for each repetition.
```

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")
```

While loops: Repeating until a condition is met

```python
While loops in Python repeat a block of code as long as a specified condition remains true. The loop checks the condition before each iteration, and 
if the condition is false, the loop stops running. This allows you to repeat tasks dynamically until a certain criterion is met, such as reaching a 
target number or receiving specific user input.While loops are useful when the number of repetitions is not known beforehand and depends on changing 
conditions during program execution.
```

```python
count = 1
while count <= 5:
    print(f"Count is: {count}")
    count += 1
```

Basic error handling: What to do when things go wrong

```python
Basic error handling in Python is done using the try and except blocks which allow your program to deal with unexpected problems gracefully 
instead of crashing. You place the code that might cause an error inside the try block. If an error occurs in this block, Python immediately 
jumps to the except block where you can define how to handle the error, such as printing a message or taking corrective action. Optionally, 
an else block can run code when no errors occur, and a finally block runs code regardless of what happened, often used for cleanup. This structure 
helps create stable programs that anticipate and manage errors like dividing by zero or invalid inputs.
```

```python
try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ZeroDivisionError:
    print("Error: You cannot divide by zero.")
except ValueError:
    print("Error: Please enter a valid number.")
else:
    print("Result is", result)
finally:
    print("Execution complete.")
```

Make a simple calculator that can add, subtract, multiply, divide

```python
def calculator():
    print("Choose an operation: +, -, *, /")
    operation = input("Operation: ")
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if operation == '+':
        print("Result:", num1 + num2)
    elif operation == '-':
        print("Result:", num1 - num2)
    elif operation == '*':
        print("Result:", num1 * num2)
    elif operation == '/':
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Error: Cannot divide by zero.")
    else:
        print("Invalid operation.")

calculator()
```

Practice with real-life scenarios: grading system, age categories

```python
score = float(input("Enter your score (0-100): "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your grade is {grade}.")
```

## Build a Simple Student Grade Calculator that takes marks and returns grade (A, B, C, etc.) with encouraging messages for each grade

```python
def grade_calculator():
    marks = float(input("Enter the marks obtained (0-100): "))

    if marks >= 90:
        grade = "A"
        message = "Excellent work! Keep it up!"
    elif marks >= 80:
        grade = "B"
        message = "Good job! You're doing well!"
    elif marks >= 70:
        grade = "C"
        message = "Fair effort, keep improving!"
    elif marks >= 60:
        grade = "D"
        message = "Needs improvement, but you can do it!"
    else:
        grade = "F"
        message = "Don't give up! Keep trying and you'll get better!"

    print(f"Your grade is {grade}. {message}")

grade_calculator()
```
