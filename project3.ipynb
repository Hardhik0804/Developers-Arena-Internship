What is data analysis? Understanding why it's important

```python
Data analysis is the systematic process of collecting, cleaning, transforming, and interpreting data to extract useful information and support 
decision-making. It helps identify patterns, trends, and relationships within data, enabling individuals and organizations to make informed choices 
and solve problems effectively. The process often involves preparing the data for analysis, exploring it with statistical or visual techniques, and 
drawing conclusions that can guide actions. Data analysis is essential across many fields like business, healthcare, education, and science because it
turns raw data into actionable insights that improve outcomes and drive progress.
```

Exploring data: seeing basic information about your dataset

```python
import pandas as pd

df = pd.read_csv('filename.csv')

print(df.head())
print(df.info())
print(df.describe())

Exploring data in Python using Pandas involves viewing basic information and summary statistics about your dataset to understand its structure and 
contents. Key functions include:
df.head() shows the first few rows of the dataset to quickly glimpse the data.
df.info() provides a concise summary, including the number of non-null values, data types of each column, and memory usage.
df.describe() calculates summary statistics like count, mean, standard deviation, minimum, maximum, and quartiles for numeric columns.
Functions like df.mean(), df.median(), and df.mode() give specific statistics for columns.
You can also check for missing values using df.isnull().sum().
These exploratory methods help you quickly assess data quality, distribution, and key insights before performing deeper analysis or modeling.
```

Data cleaning: fixing missing or wrong values

```python
import pandas as pd

df = pd.read_csv('data.csv')

# Check missing values
print(df.isnull().sum())

# Fill missing values in 'Age' column with the mean age
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Remove rows with any missing values
df.dropna(inplace=True)

print(df.info())

Data cleaning in Python, especially using Pandas, involves identifying and fixing missing or incorrect values in your dataset to ensure accuracy and 
consistency for analysis. Common techniques include:
Detecting missing values: Use df.isnull().sum() to count missing entries in each column.
Handling missing values: You can remove rows with df.dropna() or fill missing values with a specific value or strategy using df.fillna(value) or
methods like forward fill (ffill) or backward fill (bfill).
Correcting wrong values: Identify anomalies or outliers by applying filters or conditions, and then replace or remove these incorrect values.
Data type conversion: Ensure columns have appropriate types using df.astype() for consistent operations.
Effective data cleaning improves the reliability of analysis by addressing issues like incomplete or inconsistent data entries.
```

Analyze a Simple Sales Dataset to find: total sales, best-selling product, and create a basic report

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv('sales_data.csv')
```

```python
sales
```

```python
sales.shape
```

```python
sales.describe()
```

```python
sales.isnull().sum()
```

```python
# total sales
total_sales = sales["Sales_Amount"].sum()
```

```python
# Best-selling product by revenue 
best_product = (sales.groupby("Product_ID")["Sales_Amount"].sum().sort_values(ascending=False).head(1))
```

```python
best_product
```

```python
best_product_quantity = (sales.groupby("Product_ID")["Quantity_Sold"].sum().sort_values(ascending=False).head(1))
```

```python
best_product_quantity
```

```python
sales_catageroy = (sales.groupby("Product_ID")["Product_Category"].sum().sort_values(ascending=False))
```

```python
sales_catageroy
```

```python
top_products = (
    sales.groupby("Product_ID")["Sales_Amount"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))
top_products.plot(kind="bar", color="orange")
plt.title("Top 10 Products by Revenue")
plt.xlabel("Product ID")
plt.ylabel("Sales Amount")
plt.tight_layout()
plt.show()
```
