Basic math operations: addition, subtraction, multiplication, division

```python
a = 10
b = 5

addition = a + b          # Addition: 10 + 5 = 15
subtraction = a - b       # Subtraction: 10 - 5 = 5
multiplication = a * b    # Multiplication: 10 * 5 = 50
division = a / b          # Division: 10 / 5 = 2.0

print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```

Strings: Working with text data - creating, combining, and basic operations

```python
Strings in Python are used to work with text data. You can create strings by enclosing text in quotes, combine strings using the + operator for 
concatenation, and perform basic operations such as finding the length, converting case, or extracting parts of the string. Python also provides 
many built-in methods like upper(), lower(), strip(), and split() for manipulating strings efficiently. These operations help in formatting, 
searching, and managing text data conveniently within programs.
```

Lists: Collections of items and how to access them

```python
Lists in Python are ordered collections of items enclosed in square brackets, capable of holding multiple elements which can be of different data types 
such as numbers, strings, or even other lists. They are mutable, meaning you can modify, add, or remove elements after the list is created. Items in a
list are accessed using zero-based indexing, where the first item is at index 0, the second at index 1, and so forth. Python also supports negative 
indexing to access elements from the end, with -1 representing the last item. Lists are versatile data structures often used to organize and manipulate
collections of related data efficiently.
```

Your first program: Understanding how code runs step by step

```python
Your first program introduces the fundamental concept of how code executes step by step in a programming language like Python. When you write a program,
the computer reads and executes your instructions line by line from top to bottom. Each statement is processed sequentially, performing actions like 
assigning values to variables, performing calculations, or displaying outputs. For example, in a simple program that adds two numbers and shows the 
result, the computer first stores the numbers, then adds them, and finally prints out the sum. Understanding this sequential flow helps you write 
programs that perform tasks logically and predictably.
```
